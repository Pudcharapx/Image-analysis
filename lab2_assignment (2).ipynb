```python
import cv2
import numpy as np

img = cv2.imread('F1.png')

# การแยกสีแดงออกมาโดยการสับเปลี่ยนระนาบอาเรย์ชั้นที่ 2 (BGR -> Index 2)
# หมายเหตุ: การใช้ Array Slicing [:, :, 2] จะเร็วกว่าการใช้คำสั่ง cv2.split()
red_channel = img[:, :, 2]

# สร้างภาพสีดำที่มีขนาดเท่ากับภาพต้นฉบับ
blank_canvas = np.zeros_like(img)

# นำช่องสีแดงเดี่ยว ๆ ไปใส่ไว้ในแชนเนลสีแดงของภาพเปล่า เพื่อจำลองผลลัพธ์ภาพสีแดงเชิงเดี่ยว
blank_canvas[:, :, 2] = red_channel

cv2.imshow('Pure Red Channel View', blank_canvas)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# อ่านไฟล์ภาพ และ แปลงเป็นสีเทา
img_gray = cv2.imread("F1.png", cv2.IMREAD_GRAYSCALE)

# คัดลอกภาพจากต้นฉบับ
img_binary = np.zeros_like(img_gray)

# พิกเซลที่น้อยกว่า 128 ให้เป็น 0
img_binary[img_gray < 128] = 0
# พิกเซลที่มากกว่าหรือเท่ากับ 128 ให้เป็น 255
img_binary[img_gray >= 128] = 255

# บันทึกภาพ
cv2.imwrite("output.png", img_binary)

plt.figure(figsize = (10,5))

# พล้อตภาพที่ 1 
plt.subplot(1, 2, 1)
plt.imshow(img_gray, cmap='gray', vmin=0, vmax=255)
plt.title('Original Grayscale Image')
plt.axis('off')

# พล้อตภาพที่ 2
plt.subplot(1, 2, 2)
plt.imshow(img_binary, cmap='gray', vmin=0, vmax=255)
plt.title('Binary Image (Threshold = 128)')
plt.axis('off')

plt.tight_layout()
plt.show()
```

<b>🔍 คำอธิบายโค้ดส่วนนี้:</b><ul>
<li><b>แปลงสเปซสี (Color Space):</b> เปลี่ยนรูปแบบสีของภาพ (เช่น จาก BGR เป็น RGB หรือ Grayscale)</li>
<li><b>แสดงผลภาพ (Visualization):</b> ใช้ matplotlib เพื่อแสดงผลภาพหรือกราฟเปรียบเทียบ</li>
</ul></div>


```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# สร้างภาพเปล่า 500 * 500 ใช้ช่องสามสี BGR 
img = np.zeros((500 , 500 , 3), dtype = np.uint8)

# กำหนดจุดศูนย์กลางภาพ
cx , cy = 250 , 250

# สร้างกริดพิกัด x และ y สำหรับภาพขนาด 500x500
y , x = np.ogrid[:500 , :500]

# คำนวณระยะทาง Euclidean จากทุกพิกเซลไปยังจุดศูนย์กลาง
# สูตร: De = sqrt((x - cx)^2 + (y - cy)^2)
D_e = np.sqrt((x - cx)**2 + (y - cy)**2)

# กำหนดเงื่อนไข ระยะทางอยู่ระหว่าง 148 ถึง 152
circle_mask = (D_e >= 148) & (D_e <= 152)

# กำหนดสีพิกเซลที่ตรงตามเงื่อนไขเป็นสีเหลือง (BGR: 0, 255, 255)
img[circle_mask] = [0, 255, 255]

# คำนวณระยะทาง Chebyshev (Chessboard) จากทุกพิกเซลไปยังจุดศูนย์กลาง
# สูตร D8 = max(|x - cx|, |y - cy|)
D_8 = np.maximum(np.abs(x - cx), np.abs(y - cy))

# กำหนดเงื่อนไข ระยะทางอยู่ระหว่าง 78 ถึง 82
square_mask = (D_8 >= 78) & (D_8 <= 82)

# กำหนดสีพิกเซลที่ตรงตามเงื่อนไขเป็นสีน้ำเงิน (BGR: 255, 0, 0)
img[square_mask] = [255, 0, 0]

# บันทึกภาพลงดิสก์ในชื่อ distance_shapes.png
cv2.imwrite('distance_shapes.png', img)

# แปลงพื้นที่สีจาก BGR เป็น RGB เพื่อให้ Matplotlib แสดงสีได้ถูกต้อง
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# แสดงผลออกทางหน้าจอคอมพิวเตอร์
plt.figure(figsize=(6, 6))
plt.imshow(img_rgb)
plt.title('Mathematical Distance-Based Overlay\nCircle ($D_e$) & Square ($D_8$)')
plt.axis('on')
plt.show()
```
